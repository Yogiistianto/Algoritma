```python
# Penggunaan Loop untuk mengganti element di dalam list

kotak = ['merah', 'kuning', 'hijau', 'ungu', 'biru']

for i in range(0, 5):
    print("Sebelum bangun datar persegi ", i, 'adalah',  kotak[i])
    kotak[i] = 'massa'
    print("Setelah bangun datar persegi ", i, 'adalah',  kotak[i], "\n")
```

```python
# Loop untuk pengurutan dan pengisian data pada list

lingkaran=['merah', 'kuning', 'hijau', 'ungu', 'biru']

for i, bundaran in enumerate(lingkaran):
    print(i, bundaran, "\n")
```

```python
# While Loop

masa = [1982, 1980, 1973, 2000]

i = 0
tahun = 0

while(tahun != 1973):
    tahun = masa[i]
    i = i + 1
    print(tahun)

print("Butuh", i ,"pengulangan untuk keluar dari Loop.\n")
```

```python
# Nested Loop

# loop bagian luar
for i in range(1, 11):
    # nested loop
    # untuk setiap nilai i, loop ini akan dijalankan sebanyak 10 kali
    for j in range(1, 11):
        # print perkalian
        print(i * j, end=' ')
    print( "", "\n")
```

```python
# while Loop di dalam for Loop

quotes = ['Lah', 'gw', 'Mah', 'Ganteng']
# loop bagian luar
for name in quotes:
    # bagian dalam while loop
    count = 0
    while count < 10:
        print(name, end=' ')
        # penghitung increment
        count = count + 1
    print("", "\n")
```

```python
# Jeda pada nested Loop

for i in range(4):
    for j in range(4):
        if j == i:
            break
        print(i, j, "\n")
```

```python
# Exercise 2
print("membuat segitiga pascal")
n  = int(input("Masukkan angka: "))
c = 1

for i in range(1 , n+1):
  for j in range(1,(n-i)+1):
    print(" ", end="")
  for k in range(0,i):
    if(k==0 or i==0):
      c=1
    else:
      c=c*(i-k)//k
    print(c,"",end="")
  print()
```

```python
#Exercise 3
print("Menampilkan jumlah angka dalam satu baris segitiga pascal")
jumlah_baris = int(input("Masukkan jumlah baris: "))

segitiga_pascal = []
for i in range(jumlah_baris):
    baris = []
    for j in range(i + 1):
        if j == 0 or j == i:
            baris.append(1)
        else:
            baris.append(segitiga_pascal[i - 1][j - 1] + segitiga_pascal[i - 1][j])
    segitiga_pascal.append(baris)

if jumlah_baris > 0:
    print("Segitiga Pascal:")
    last_angka_baris = segitiga_pascal[jumlah_baris - 1]
    print("Angka dalam sebaris segitiga pascal ", end=' ')
    for angka in last_angka_baris:
        print(angka, end=' ')
    print()
    jumlah_angka = 0
    for angka in last_angka_baris:
        jumlah_angka += angka
    print("Jumlah angka dalam baris terakhir:", jumlah_angka)
else:
    print("Jumlah baris harus lebih besar dari 0.")
```
