```python
# LIST

# Kumpulan atau tipe data yang berurutan dan dapat diubah. Elemen dalam list dapat di duplikat
# Penulisan list manggunakan tanda []

thislist = ["apple", "banana", "cherry"]
print(thislist)
```

```python
# TUPLE

# Kumpulan atau tipe data yang berurutan, namun tidak dapat diubah.Elemen dalam tuple dapat di duplikat
# Penulisannya menggunakan tanda ()

thistuple = ("apple", "banana", "cherry")
print(thistuple)
```

```python
# SET

# Kumpulan atau tipe data yang tidak berurutan, tidak dapat diubah, dan tidak terindeks. Elemen dalam tipe data set tidak dapat di duplikat
# Penulisana set menggunakan tanda {}

thisset = {"apple", "banana", "cherry"}
print(thisset)
```

```python
# DICTONARY

# Kumpulan atau tipe data yang berurutan, dan dapat diubah. Elemen dalam dictonary tidak dapat di duplikat
# Penulisan menggunakan tanda {}, dengan setiap pasangan kunci nilai dipisahkan oleh titik dua (:)

thisdict = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964
}
print(thisdict)
```

```python
# WEEKLY EXERCISE 2
print("WEEKLY EXERCISE 2\n")

# Tipe collection untuk mengelola data

# Membuat dictionary untuk menyimpan data mahasiswa
mahasiswa1 = {
    "nama": "Muhammad Yogi Istianto Wibowo",
    "kelas": "Teknologi Informasi Rombel 1",
    "tanggal_lahir": "23, Juni 1976",
    "no_hp": "081228895564",
    "alamat": "Jl. Soedirman No. 6, Magelang",
    "hobi": ["Bermain Game", "Mengaji", "Membaca", "Olahraga"]
}

# Membuat list untuk menyimpan data mahasiswa
daftar_mahasiswa = [
    mahasiswa1,
]

# Menampilkan data mahasiswa
for x in daftar_mahasiswa:
    print("Nama           :", x["nama"])
    print("Kelas          :", x["kelas"])
    print("Tanggal Lahir  :", x["tanggal_lahir"])
    print("No HP          :", x["no_hp"])
    print("Alamat         :", x["alamat"])
    print("Hobi           :")
    for hobi in x["hobi"]:
        print("-", hobi)
    print()
```
