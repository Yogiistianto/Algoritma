```python
# Kamus data mahasiswa (dictionary)
mahasiswa = {
    "Shafira" : [85, 75],
    "Amanda" : [80, 90],
    "Aditya" : [75, 80],
    "Nedia" : [95, 80],
    "Widya" : [85, 85],
    "Hanif" : [75, 90],
    "Andi" : [70, 75],
    "Dhanar" : [85, 85],
    "Hikma" : [80, 75]
}
def hitung_rata_rata(nilai):
    return sum(nilai) / len(nilai)
# Indent the code block below to fix the error
if __name__ == "__main__":
    rata_rata_shafira = hitung_rata_rata(mahasiswa["Shafira"])
    print("Rata-rata Shafira:", rata_rata_shafira)
    jumlah_nilai_hanif_andi = sum(mahasiswa["Hanif"]) + sum(mahasiswa["Andi"])
    print("Jumlah nilai Hanif dan Andi:", jumlah_nilai_hanif_andi)
    mahasiswa_terpilih = ["Widya", "Dhanar", "Hikma", "Nedia"]
    nilai_kalkulus = [mahasiswa[m][0] for m in mahasiswa_terpilih]
    nilai_statistika = [mahasiswa[m][1] for m in mahasiswa_terpilih]
    rata_rata_kalkulus = hitung_rata_rata(nilai_kalkulus)
    rata_rata_statistika = hitung_rata_rata(nilai_statistika)
    print("Rata-rata nilai kalkulus:", rata_rata_kalkulus)
    print("Rata-rata nilai statistika:", rata_rata_statistika)
    nilai_kalkulus_semua = [nilai [0] for nilai in mahasiswa.values()]
    rata_rata_kalkulus_semua = hitung_rata_rata(nilai_kalkulus_semua)
    print("Rata-rata nilai kalkulus 1 untuk semua mahasiswa:", rata_rata_kalkulus_semua)
    nilai_stastika_semua = [nilai [1] for nilai in mahasiswa.values()]
    rata_rata_stastika_semua = hitung_rata_rata(nilai_stastika_semua)
    print("Rata-rata nilai statistika 1 untuk semua mahasiswa:", rata_rata_stastika_semua)
```

```python
# Konstanta
JAM_KERJA_SEHARI = 8
UANG_JAM = 30000
TUNJANGAN_PERSEN = 10
UANG_LEMBUR_PER_JAM = 10000

# Input jumlah hari tidak masuk kerja di bulan Agustus
hari_tidak_masuk = int(input("3 : "))

# Hitung jumlah hari kerja di bulan Agustus (asumsi 31 hari)
hari_kerja = 31 - hari_tidak_masuk

# Hitung total jam kerja
total_jam_kerja = hari_kerja * JAM_KERJA_SEHARI

# Hitung gaji pokok
gaji_pokok = total_jam_kerja * UANG_JAM

# Hitung tunjangan (jika berlaku)
tunjangan = gaji_pokok * (TUNJANGAN_PERSEN / 100) if hari_kerja >= (31 * 5 / 12) else 0

# Input jumlah jam lembur (opsional)
jam_lembur = int(input("Masukkan jumlah jam lembur di bulan Agustus (0 jika tidak ada): "))

# Hitung gaji lembur
gaji_lembur = jam_lembur * UANG_LEMBUR_PER_JAM

# Hitung total gaji
total_gaji = gaji_pokok + tunjangan + gaji_lembur

# Tampilkan hasil
print("Gaji pokok:", gaji_pokok)
print("Tunjangan:", tunjangan)
print("Gaji lembur:", gaji_lembur)
print("Total gaji:", total_gaji)
```
