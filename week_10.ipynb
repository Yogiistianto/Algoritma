```python
mediudef faktorial(n):
    if n == 0 or n == 1:
        return 1
    else:
        return n * faktorial(n-1)

def main():
    try:
        bilangan = int(input("Masukkan bilangan untuk menghitung faktorial: "))

        if bilangan < 0:
            print("Maaf, faktorial tidak bisa dihitung untuk bilangan negatif")
            return

        hasil = faktorial(bilangan)

        print(f"Faktorial dari {bilangan} adalah: {hasil}")

    except ValueError:
        print("Mohon masukkan bilangan bulat yang valid")

if __name__ == "__main__":
    main()
```

```python
import math

def hitung_luas_lingkaran():
    try:
        jari_jari = float(input("Masukkan jari-jari lingkaran: "))
        if jari_jari < 0:
            print("Error: Jari-jari tidak boleh negatif!")
            return
        luas = math.pi * jari_jari ** 2
        print(f"\nLuas lingkaran dengan jari-jari {jari_jari} adalah: {luas:.2f}")
    except ValueError:
        print("Error: Mohon masukkan angka yang valid!")

def hitung_luas_persegi():
    try:
        sisi = float(input("Masukkan panjang sisi persegi: "))
        if sisi < 0:
            print("Error: Panjang sisi tidak boleh negatif!")
            return
        luas = sisi ** 2
        print(f"\nLuas persegi dengan sisi {sisi} adalah: {luas:.2f}")
    except ValueError:
        print("Error: Mohon masukkan angka yang valid!")

def hitung_luas_segitiga():
    try:
        alas = float(input("Masukkan panjang alas segitiga: "))
        tinggi = float(input("Masukkan tinggi segitiga: "))
        if alas < 0 or tinggi < 0:
            print("Error: Panjang alas dan tinggi tidak boleh negatif!")
            return
        luas = 0.5 * alas * tinggi
        print(f"\nLuas segitiga dengan alas {alas} dan tinggi {tinggi} adalah: {luas:.2f}")
    except ValueError:
        print("Error: Mohon masukkan angka yang valid!")

def main():
    while True:
        print("\nKalkulator Luas Bangun Datar\n")
        print("1. Hitung Luas Lingkaran")
        print("2. Hitung Luas Persegi")
        print("3. Hitung Luas Segitiga")
        print("4. Keluar")

        pilihan = input("\nPilih menu (1-4): ")

        if pilihan == "1":
            hitung_luas_lingkaran()
        elif pilihan == "2":
            hitung_luas_persegi()
        elif pilihan == "3":
            hitung_luas_segitiga()
        elif pilihan == "4":
            print("Terima kasih telah menggunakan kalkulator ini!")
            break
        else:
            print("Error: Pilihan tidak valid! Silakan pilih 1-4.")

        lanjut = input("\nApakah ingin melakukan perhitungan lain? (y/n): ")
        if lanjut.lower() != 'y':
            print("Terima kasih telah menggunakan kalkulator ini!")
            break

if __name__ == "__main__":
    main()
```
